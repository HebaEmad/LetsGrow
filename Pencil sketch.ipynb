```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import cv2
```

```python
img = cv2.imread('Traumatized_Mr._Incredible.jpg')
```

```python
img = cv2.cvtColor(img,cv2.COLOR_BGR2RGB)
plt.axis('off')
plt.imshow(img)
plt.show()
```

## Convert the image to grayscale

```python
gray_img = cv2.cvtColor(img, cv2.COLOR_BGR2GRAY)
plt.imshow(cv2.cvtColor(gray_img, cv2.COLOR_BGR2RGB))
plt.axis('off')
plt.show()
```

## Inverted gray image

```python
inverted_img = 255 - gray_img
plt.imshow(cv2.cvtColor(inverted_img, cv2.COLOR_BGR2RGB))
plt.axis('off')
plt.show()
```

## Blur the image by using the Gaussian Function

```python
blurred_img = cv2.GaussianBlur(inverted_img, (21, 21), 0)
pencil_sketch = cv2.divide(gray_img, blurred_img, scale=256.0)

# Convert the pencil sketch back to RGB for display
pencil_sketch_rgb = cv2.cvtColor(pencil_sketch, cv2.COLOR_GRAY2RGB)

# Display the pencil sketch
plt.imshow(pencil_sketch_rgb)
plt.axis('off')
plt.show()
```
